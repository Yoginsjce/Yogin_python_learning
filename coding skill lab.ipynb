```python
def second_largest(arr):
    arr = list(set(arr))  # Remove duplicates
    arr.sort()  # Sort the array
    return arr[-2]  # Second last element

print(second_largest([10, 102, 45, 99, 77,143]))
```

```python
def reverse_string(s):
    rev = ""
    for char in s:
        rev = char + rev
    return rev

print(reverse_string("yogin"))
```

```python
def find_missing(arr, n):
    return sum(range(1, n + 1)) - sum(arr)

print(find_missing([1, 2, 4, 5], 5))  # Output: 3
```

```python
def longest_common_prefix(words):
    prefix = words[0]
    for word in words:
        while not word.startswith(prefix):
            prefix = prefix[:-1]
    return prefix

print(longest_common_prefix(["flower", "flow", "flight","finger"]))
```

```python
def first_non_repeating(s):
    for char in s:
        if s.count(char) == 1:
            return char
    return None

print(first_non_repeating("swwiiss"))
```

```python
def binary_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

print(binary_search([1, 3, 5, 7, 9], 7))
```

```python
def kth_smallest(arr, k):
    return sorted(arr)[k - 1]

print(kth_smallest([7, 10, 4, 3, 20, 15], 4))  # Output: 7
```

```python
def majority_element(arr):
    for num in arr:
        if arr.count(num) > len(arr) // 2:
            return num
    return -1

print(majority_element([3, 3, 4, 2, 3, 3, 3]))
```

```python
def intersection(arr1, arr2):
    return [num for num in arr1 if num in arr2]

print(intersection([1, 2, 4, 5, 6], [2, 3, 5, 7,1]))
```
